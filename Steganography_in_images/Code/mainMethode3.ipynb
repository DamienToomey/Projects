# Méthode 3 : LSB : Least Significant Bit
# Pour chaque caractère du message, on met chaque bit de ce caractère dans le bit de poids faible de chaque couleur des trois pixels consécutifs donnés par la fonction de position

# Imports

```python
from numpy import uint8
from numpy import zeros
from numpy import unpackbits # conversion uint8 en binaire
# Exemple : 63 uint8 --unpackbits--> 00111111 en binaire
from numpy import packbits # conversion binaire en uint8
# Exemple :  00111111 en binaire --packbits--> 63 uint8

import matplotlib.pyplot as plt

# sudo pip3 install imageio
from imageio import imread # lire l'image en uint8
from imageio import imwrite # ecrire l'image dans un fichier

# Pour utiliser des fonctions d'autres fichiers jupyter
# sudo pip3 install ipynb
from ipynb.fs.full.mesFonctions import initialisation
```

# Image Originale

```python
nomImage = "windowsXP.png"
[image,n,p,z,pathDonnees,pathResultats] = initialisation(nomImage)
```

# Choix du message

```python
message = "Il fait beau aujourd'hui"
longueurMessage = len(message)
```

# Choix des fonctions de position

```python
def fonctionHachageX(k):
    return 5*k+10
#enddef

def fonctionHachageY(k):
    return 2*k+20
#enddef
```

# Codage du message dans l'image

```python
pythonOblige = 1

for k in range(0,longueurMessage,1):
    
    posX = fonctionHachageX(k)
    posY = fonctionHachageY(k)
    
    if ((posX<n-2) & (posY<p-2)):    
        codeAscii=uint8(ord(message[k]))        
        codeAsciiEnBinaire = unpackbits(codeAscii) # uint8 en 8 bits
        
        indice = 0
        for l in range (0,3,1):
            
            pixelREnBinaire = unpackbits(image[posX+l,posY,0]) # uint8 en 8 bits
            pixelREnBinaire[7] = codeAsciiEnBinaire[indice]
            image[posX+l,posY,0] = packbits(pixelREnBinaire)
            indice = indice+1
            
            pixelVEnBinaire = unpackbits(image[posX+l,posY,1]) # uint8 en 8 bits
            pixelVEnBinaire[7] = codeAsciiEnBinaire[indice]
            image[posX+l,posY,1]=packbits(pixelVEnBinaire)
            indice = indice+1
            
            if (indice<len(codeAsciiEnBinaire)):
                pixelBEnBinaire = unpackbits(image[posX+l,posY,2]) # uint8 en 8 bits
                pixelBEnBinaire[7] = codeAsciiEnBinaire[indice]
                image[posX+l,posY,2] = packbits(pixelBEnBinaire)
                indice = indice+1
            #endif
        #endfor
    else:
        raise Exception("Le message sort de l'image")
#endfor
```

# Sauvegarde de l'image

```python
nomImageAvecMessage = "resultatMethode3.png"
imwrite(pathResultats+nomImageAvecMessage, image)
```

```python
message=""
codeAsciiEnBinaire = zeros(8).astype(uint8)

for k in range(0,longueurMessage,1):
    posX = fonctionHachageX(k)
    posY = fonctionHachageY(k)
    
    indice = 0
    for l in range (0,3,1):
        pixelREnBinaire = unpackbits(image[posX+l,posY,0]) # uint8 en 8 bits
        codeAsciiEnBinaire[indice] = pixelREnBinaire[7]
        indice = indice+1        

        pixelVEnBinaire = unpackbits(image[posX+l,posY,1]) # uint8 en 8 bits
        codeAsciiEnBinaire[indice] = pixelVEnBinaire[7]
        indice = indice+1
            
        if (indice<len(codeAsciiEnBinaire)):
            pixelBEnBinaire = unpackbits(image[posX+l,posY,2]) # uint8 en 8 bits
            codeAsciiEnBinaire[indice] = pixelBEnBinaire[7]
            indice = indice+1
        #endif
    #endfor
    
    message=message+chr(packbits(codeAsciiEnBinaire))
#endfor

print(message)
```
