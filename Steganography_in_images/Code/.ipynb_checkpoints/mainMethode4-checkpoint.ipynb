# Méthode 4 : ELSB : EDGE LEAST SIGNIFICANT BIT EMBEDDING

# Pour chaque caractère du message, pour chaque paire de bits de ce caractère, je mets cette paire dans les deux bits de poids faible du rouge de 4 pixels dans l'image originale aux coordonnées des contours de l'image masquée

# Imports

```python
from numpy import uint8
from numpy import zeros
from numpy import unpackbits # conversion uint8 en binaire
# Exemple : 63 uint8 --unpackbits--> 00111111 en binaire
from numpy import packbits # conversion binaire en uint8
# Exemple :  00111111 en binaire --packbits--> 63 uint8
from numpy import copy # copier le contenu d'un array
from numpy import array
from numpy import unique
from numpy import where

import matplotlib.pyplot as plt

# sudo pip3 install imageio
from imageio import imread # lire l'image en uint8
from imageio import imwrite # ecrire l'image dans un fichier

# Pour utiliser des fonctions d'autres fichiers jupyter
# sudo pip3 install ipynb
from ipynb.fs.full.mesFonctions import initialisation
from ipynb.fs.full.monCannyDeriche import mainCannyDeriche

# pip3 install opencv-python
import cv2 as cv
```

# Initialisation

```python
nomImage = "office.png"
[image,n,p,z,pathDonnees,pathResultats] = initialisation(nomImage)
```

# Choix du message

```python
message = "Il fait beau aujourd'hui"
longueurMessage = len(message)
```

# Masquage des deux bits de poids faible de chaque pixel

```python
def monMasque(image,n,p,z):
    pythonOblige = 1
    
    imageMasked = copy(image)

    for i in range (0,n,1):
        for j in range (0,p,1):
            for k in range (0,z,1):
                bitsArray = unpackbits(imageMasked[i,j,k])
                bitsArray[6:7+pythonOblige] = array([0,0]).astype(uint8)
                imageMasked[i,j,k] = packbits(bitsArray)
            #endfor
        #endfor
    #endfor
    
    return imageMasked
#enddef
```

```python
imageMasked = monMasque(image,n,p,z)

plt.figure(figsize = (50,10))
plt.imshow(imageMasked)
plt.title('Image Originale Masquée'), plt.xticks([]), plt.yticks([])
plt.show()
```

# Détecteur de contours de Deriche

## Choix des seuils pour le seuillage par hystérisis

```python
seuil_min = 10 # pour openCV 100 
seuil_max = 15 # pour openCV 200
```

```python
def Deriche(imageMasked,seuil_min,seuil_max):
    imgageMaskedGray = cv.cvtColor(imageMasked, cv.COLOR_BGR2GRAY) # convertir l'image en niveaux de gris

    # RGB to gray : RGB[A] to Gray:Y←0.299⋅R+0.587⋅G+0.114⋅B

    [nGray,pGray] = imgageMaskedGray.shape

    if (len(imgageMaskedGray.shape)==2):
        zGray = 1
    #endif

    print("Dimensions imageGray : nGray =", nGray, "; pGray =", pGray, "; zGray =", zGray)
    
    # Canny-Deriche fait par moi :
    alpha = 1 
    omega = 0.0001
    edges = mainCannyDeriche(imgageMaskedGray,alpha,omega,seuil_min,seuil_max)
    
    # Open CV :
    #edges = cv.Canny(imgageMaskedGray,seuil_min,seuil_max) # seuil_min, seuil_max
    
    plt.figure(figsize = (50,10))
    plt.imshow(edges,cmap = 'gray')
    plt.title("Contours de l'Image"), plt.xticks([]), plt.yticks([])

    plt.show()

    return edges
#enddef
```

```python
edges = Deriche(imageMasked,seuil_min,seuil_max)
```

# Valeurs résultantes dans la variable $ edges $

```python
unique(edges)
```

# Codage du message dans l'image

```python
[posX,posY]=where(edges==255)
pythonOblige = 1

k = 0
it = 0

while (k<longueurMessage):
    
    if ((posX[it]<len(posX)-3) & (posY[it]<len(posY)-3)):   
        
        codeAscii=uint8(ord(message[k]))        
        codeAsciiEnBinaire = unpackbits(codeAscii) # uint8 en 8 bits
        
        indice = 0
        
        for l in range (it,it+4,1): # parcourir 4 pixels
            
            pixelREnBinaire = unpackbits(image[posX[l],posY[l],0]) # uint8 en 8 bits
            pixelREnBinaire[6:7+pythonOblige] = codeAsciiEnBinaire[indice:indice+1+pythonOblige]
            image[posX[l],posY[l],0] = packbits(pixelREnBinaire)
            
            indice = indice+2            
        #endfor
    else:
        raise Exception("Le message sort de l'image")
    it = l+1
    k = k+1
#endwhile
```

# Sauvegarde de l'image

```python
nomImageAvecMessage = "resultatMethode4.png"
imwrite(pathResultats+nomImageAvecMessage, image)
```

# Extraction du message dans l'image

```python
[posX,posY]=where(edges==255)
pythonOblige = 1

message=""
codeAsciiEnBinaire = zeros(8).astype(uint8)

k = 0
it = 0

while (k<longueurMessage):
    
    indice = 0    
 
    for l in range (it,it+4,1): # parcourir 4 pixels
        pixelREnBinaire = unpackbits(imageAvecMessage[posX[l],posY[l],0]) # uint8 en 8 bits
        codeAsciiEnBinaire[indice:indice+1+pythonOblige] = pixelREnBinaire[6:7+pythonOblige]
            
        indice = indice+2
    #endfor
    
    message=message+chr(packbits(codeAsciiEnBinaire))
    it = l+1
    k = k+1
#endwhile

print(message)
```
